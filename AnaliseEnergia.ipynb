```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
BaseDados = pd.read_excel('Dados Empresas Energia.xlsx')
BaseDados.head()
```

```python
BaseDados.set_index('Data', inplace=True)
BaseDados.head()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(16,7))
plt.title('Análise ações de empresas de Energia', loc='left', fontsize=18)
plt.plot(BaseDados.index, BaseDados['Petrobras'], color='green', linewidth=4, alpha=0.7)
plt.text(BaseDados.index[-1], BaseDados['Petrobras'].tail(1), 'Petrobras', color='red', size='large', horizontalalignment='left')

for Coluna in BaseDados.columns[1:]:
    plt.plot(BaseDados.index, BaseDados[Coluna], color='gray', linewidth=2, alpha=0.7)
    plt.text(BaseDados.index[-1], BaseDados[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left')

plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
